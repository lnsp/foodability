```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("food_data.csv")
```

```python
df
```

```python
tags = list(df["packaging_tags"].unique())
tags
```

```python
PLASTIC_TAGS = ["plastic", "plastique", "plastico"]

def process_packaging_tags(tag_str):
    try:
        results = set()
        tag_list = tag_str.split(",")
        for tag in tag_list:
            if any([t in tag for t in PLASTIC_TAGS]):
                results.add("plastic")
        return list(results)
    except Exception:
        return []
```

```python
process_packaging_tags(tags[10])
```

```python
class FoodItem:
    """Food item class"""
    
    def __init__(self, row: pd.Series):
        self.name = row["product_name"].lower()
        self.url = row["image_url"].lower()
        self.quantity = None if math.isnan(row["quantity"]) else row["quantity"]
        self.packaging_tags = process_packaging_tags(row["packaging_tags"])
```

```python
row = df.iloc[1]
```
